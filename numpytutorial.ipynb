```python
import numpy as np

a = np.array([1,2,3])
print(a)

b = np.array([[9.0, 8.0, 7.0],
              [6.0, 5.0, 4.0]])
print(b)
print(b.ndim)
```

```python
#All 0s matix
np.zeros((2,3))
```

```python
#Any other number
np.full((2,2),99)
```

```python
#Any other number (full_like)
np.full_like(a,4)
```

```python
#Random Decimal Number
np.random.random_sample(a.shape)
```

```python
#Random integer values
np.random.randint(7, size=(3,3))
```

```python
np.identity(5)
```

```python
#Repeat an array
arr = np.array([1,2,3])
r1 = np.repeat(arr,3, axis=0)
print(r1)
```

```python
output = np.ones((5,5))
print(output)

z = np.zeros((3,3))
z[1,1] = 9
print(z)

output[1:4,1:4] = z
print(output)
```

# Be Careful When Copying arrays

```python
a = np.array([1,2,3])
b=a.copy()
b[0] = 100
print(a)
```

```python
#Take the sin
np.sin(a)
```

####LINEAR ALGEBRA

```python
a = np.ones((2,3))
print(a)

b = np.full((3,2),2)
print(b)

np.matmul(a,b)
print(a)

b = np.full((3,2),2)
print(b)

np.matmul(a,b)
```

```python
c = np.identity(3)
np.linalg.det(c)
```
